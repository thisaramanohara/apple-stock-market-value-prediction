```python
import pandas as pd

dataset=pd.read_csv('AAPL_data.csv').values
```

```python
open_value=dataset[:,1]
```

```python
from matplotlib import pyplot as plt

plt.plot(open_value)
plt.xlabel('date')
plt.ylabel('open_value')
plt.show()
```

```python
open_value=open_value.reshape(-1,1)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler((0,1))
open_value_scaled=scaler.fit_transform(open_value)
```

```python
data=[]
target=[]
```

```python
fixed_window_size=50
```

```python
for i in range(0,open_value.shape[0]-fixed_window_size):
    data.append(open_value_scaled[i:i+fixed_window_size])
    target.append(open_value_scaled[i+fixed_window_size])
```

```python
import numpy as np

data=np.array(data)
target=np.array(target)
```

```python
np.save('data',data)
np.save('target',target)
```

```python
#to train
train_data=data[:1000]
train_target=target[:1000]

#to test
test_data=data[1000:]
test_target=target[1000:]
```

```python
shifted_x_value=np.arange(1000,1209,1)

plt.plot(train_target,'r')
plt.plot(shifted_x_value,test_target,'b')
plt.xlabel('date')
plt.ylabel('open_value')
plt.show()
```

```python
from keras.models import Sequential
from keras.layers import LSTM,Dense,Dropout

model=Sequential()
model.add(LSTM(128,return_sequences=True,input_shape=data.shape[1:]))
model.add(Dropout(0.2))
model.add(LSTM(128,return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(128,return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(128,return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(1,activation='linear'))

model.compile(loss='mse',optimizer='adam')
```

```python
model.fit(train_data,train_target,epochs=200,validation_split=0.1)
```

```python
predicted_target=model.predict(test_data)

from sklearn.metrics import r2_score

accuracy=r2_score(test_target,predicted_target)
print(accuracy)
```

```python
plt.plot(test_target,'b',label='test')
plt.plot(predicted_target,'g',label='predicted')
plt.xlabel('date')
plt.ylabel('open_value')
plt.legend()
plt.show()
```
